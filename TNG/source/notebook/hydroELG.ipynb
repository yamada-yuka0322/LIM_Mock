```python
import os
import numpy as np
import h5py
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
cat_dir = "/Volumes/Extreme Pro/database/LIM_Mock/hydroELG/"
with h5py.File(os.path.join(cat_dir, "Pegase_TNG300_snap99.hdf5"), "r") as f:
    lines = f['Lines']
    print(lines.keys())
    print(f['Subhalos'].keys())
    L_HA = lines['HA luminosity'][:]
    SFR = f['Subhalos']['SFR'][:]

_ = plt.hist2d(np.log10(L_HA), np.log10(SFR), bins=100, range=[[41.5, 42.5], [-3.0, 3.0]], norm=LogNorm(), cmap="Blues", rasterized=True)
plt.xlabel(r"$\log L_{\mathrm{H}\alpha} \, (\mathrm{erg}/\mathrm{s})$", fontsize=15)
plt.ylabel(r"$\log \mathrm{SFR} \, (M_\odot \, \mathrm{yr}^{-1})$", fontsize=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.colorbar()
plt.grid()
```
